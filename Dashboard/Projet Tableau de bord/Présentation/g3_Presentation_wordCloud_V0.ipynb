#### Import libraries

```python
########## Module import ##########

# Fichiers
import json
import pandas as pd
from os import path

# Math
import numpy as np

# Visualisation
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

####  Récupération des données

```python
# Load data
with open(r"C:\Users\Stior\Documents\cours\M1SID\S2\Dashboard\code\clean_articles_final.json", "r") as f:
    data = json.load(f)
```

#### Calcul des frequences des mots-clefs

```python
kw = {}
for i, row in data.items():
    for word in row["keyword"]:
        try:
            kw[word.lower()] += 1
        except:
            kw[word.lower()] = 1
```

#### Affichage des n mot-clefs les plus présent

```python
n = 20
sorted(kw.items(), key=lambda x: x[1])[-n:]
```

#### Visualisation des mot-clefs les plus frequents 

```python
mask = np.array(Image.open("ovale_mask.png")) # Forme ovale
wc = WordCloud(background_color="white", mask=mask, width=1000, height=1000, contour_width=0, colormap="winter")
wc.generate_from_frequencies(kw)
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

#### Sauvegarde du nuage de mots

```python
# save file
wc.to_file("wcWinter.png")
```
