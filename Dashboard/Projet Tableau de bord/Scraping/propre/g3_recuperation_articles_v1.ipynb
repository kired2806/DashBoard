#### Import libraries

```python
########## Module import ##########

# Fichiers
import json

# Scraping
from requests import get
from bs4 import BeautifulSoup

# Format
import time
```

## Partie Scrapping

### Parametres

```python
NB_PAGES_MAX : int = 300 #Nombre de pages scrappées (il y a 20 résultats par pages)
equation : str = "(internet or online) and (scam or fraud)" #Equation de recherche
#Différentes parties de l'url pour les résultats de recherche
search_page : str = "https://link.springer.com/search/page/"
query : str = "?query="
date_suffix : str = "&date-facet-mode=between&facet-start-year=2000&previous-start-year=1901&facet-end-year=2021&previous-end-year=2021"
```

### Fonctions

```python
def conversion_equation(str_equation : str) -> str:
    """Documentation 
    Paramètres: 
        str_equation: Equation de recherche au format String
    Sortie: 
        url: String correspondant à l'équation de recherche dans l'url de la page de recherche   
    """
    url = str_equation.replace("(","%28")
    url = url.replace(")","%29")
    url = url.replace(" ","+")
    return url
```

```python
def get_liste_pages(search_page : str,query: str,equation: str,nb_pages : int ,date_suffix: str)-> list:
    """Documentation
     Paramètres: 
        search_page: Partie correspondant à l'url de la page de recherche
        query: Partie précédent la requête fait sur le site (équation de recherche)
        equation: Equation de recherche utilisée
        nb_pages: Le nombre de pages à scrapper
        date_suffix: Partie de l'url indiquant la restriction de date
    Sortie: 
        liste_pages: Liste des urls des pages de résultats de recherche
    """
    liste_pages = []
    for i in range(nb_pages):
        url_page = search_page + str(i+1) + query + conversion_equation(equation) + date_suffix
        liste_pages.append(url_page)
    return liste_pages
```

```python
def get_liste_articles(url_page : str) -> list:
     """Documentation
     Paramètres: 
        url_page: URL de la page dont ont veut extraire le lien des articles
    Sortie: 
        liste_articles: Liste des urls des résultats sur la page
    """
    liste_articles = []
    req = get(url_page)
    html_soup = BeautifulSoup(req.text, 'html.parser')
    liste_div_article = html_soup.find_all('a',{'class':'title'})
    for div_article in liste_div_article:
        if div_article is not None:
            url_article = "https://link.springer.com" + div_article['href']
            liste_articles.append(url_article)
    return liste_articles
```

### Constitution de la liste des liens

```python
liste_pages = get_liste_pages(search_page,query,equation,NB_PAGES_MAX,date_suffix)
liste_article_complete = []
for page in liste_pages:
    print(page)
    liste_artice_page = get_liste_articles(page)
    liste_article_complete.extend(liste_artice_page)
print(len(liste_article_complete))

#articles_url_source  
```

## Sauvegrade de la liste des articles

```python
with open('articles_url_source.json', 'w') as f:
    json.dump(liste_article_complete, f)
```
