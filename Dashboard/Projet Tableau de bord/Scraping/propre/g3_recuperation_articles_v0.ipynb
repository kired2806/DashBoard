## Partie Scrapping

### Parametres

```python
# Paramètres
NB_PAGES_MAX = 300
equation = "(internet or online) and (scam or fraud)"
url_site = "https://link.springer.com/search/page/"
url_requete = "?query="
suffixe_date = "&date-facet-mode=between&facet-start-year=2000&previous-start-year=1901&facet-end-year=2021&previous-end-year=2021"
```

### Fonctions

```python
def conversion_equation(str_equation):
    url = str_equation.replace("(","%28")
    url = url.replace(")","%29")
    url = url.replace(" ","+")
    return url
```

```python
def get_liste_pages(url_site,url_requete,equation,nb_pages,suffixe_date):
    liste_pages = []
    for i in range(nb_pages):
        url_page = url_site + str(i+1) + url_requete + conversion_equation(equation) + suffixe_date
        liste_pages.append(url_page)
    return liste_pages
```

```python
def get_liste_articles(url_page):
    liste_articles = []
    req = get(url_page)
    html_soup = BeautifulSoup(req.text, 'html.parser')
    liste_div_article = html_soup.find_all('a',{'class':'title'})
    for div_article in liste_div_article:
        if div_article is not None:
            url_article = "https://link.springer.com" + div_article['href']
            liste_articles.append(url_article)
    return liste_articles
```

### Constitution de la liste

```python
liste_pages = get_liste_pages(url_site,url_requete,equation,NB_PAGES_MAX,suffixe_date)
liste_article_complete = []
for page in liste_pages:
    liste_artice_page = get_liste_articles(page)
    liste_article_complete.extend(liste_artice_page)
with open('articles_url_source.json', 'w') as f:
    json.dump(liste_article_complete, f)
#articles_url_source  
```
