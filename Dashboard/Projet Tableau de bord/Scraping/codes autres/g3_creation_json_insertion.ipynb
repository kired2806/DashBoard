#### Import libraries

```python
########## Module import ##########

# Fichiers
import json
```

#### Récupération des données à traiter

```python
with open('clean_articles_final.json',encoding="utf-8") as json_data:
    data_dict = json.load(json_data)
    
with open('themes.json',encoding="utf-8") as json_data:
    theme_dict = json.load(json_data)
```

```python
i = 0
for key in data_dict.keys():
    data_dict[key]['theme'] = theme_dict[str(i)]
    i += 1
    
```

# Sauvegrade des données 

```python
with open('insertion_articles.json', 'w',encoding="utf-8") as f:
    json.dump(data_dict, f,ensure_ascii=False)
```
