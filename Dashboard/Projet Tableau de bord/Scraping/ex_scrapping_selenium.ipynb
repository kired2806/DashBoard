```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import Select
import time
from bs4 import BeautifulSoup
import json
import webbrowser
```

```python
chr_options = Options()
chr_options.add_argument("--window-size=800,800")

browser = webdriver.Chrome(
    executable_path="/usr/lib/chromium-browser/chromedriver/chromedriver.exe",
    #chrome_options=chr_options,
)
url = "https://www.cairn.info/"
browser.get(url)
```

```python
# Acces au site

eq_search = "arnaque en ligne OU cybercriminalité OU dropshipping OU escroqueries en ligne OU Cyberfraude"

def eq_to_cairn_search(eq_search):
    return "https://www.cairn.info/resultats_recherche.php?searchTerm=" + eq_search.replace(" ", "+")

url = eq_to_cairn_search(eq_search)
browser.get(url)
```

```python
# Récupération des liens des articles sur une page

soup = BeautifulSoup(browser.page_source, 'html.parser')
art_list = [i for i in soup.find("div",{"class":"filter-content"}).find_all("div") if "article-list-item" in i.get_attribute_list("class")]
print("nb art on page:", len(art_list))

count = 0
art_readable_url = []
for i in art_list:
    get = False
    for j in i.find("div", {"class":"media-footer"}).find_all("a"):
        button = j.find("span", {"class":"name"})
        if button != None:
            if button.text == "Résumé" or button.text == "Consulter":
                get = True
    if get:
        art_readable_url.append(i.find("div", {"class":"media-body"}).findChild("a").get_attribute_list("href")[0])
    count += 1

art_readable_url
```
