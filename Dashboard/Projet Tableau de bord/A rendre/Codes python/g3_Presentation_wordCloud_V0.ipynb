#### Description du notebook 
- **V0**: Affichage d'un nuage de mots à partir des mot-clefs les plus frequents dans nos données ( Page de garde du rapport) 

#### Calcul des frequences des mots-clefs

```python
kw : dict = {}
for i, row in data.items():
    for word in row["keyword"]:
        try:
            kw[word.lower()] += 1
        except:
            kw[word.lower()] = 1
```

#### Affichage des n mot-clefs les plus présent

```python
n : int = 20
sorted(kw.items(), key=lambda x: x[1])[-n:]
```

#### Visualisation des mot-clefs les plus frequents 

```python
mask  = np.array(Image.open("ovale_mask.png")) # Forme ovale
wc = WordCloud(background_color="white", mask=mask, width=1000, height=1000, contour_width=0, colormap="winter")
wc.generate_from_frequencies(kw)
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
